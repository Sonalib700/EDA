```python
ax = sns.countplot(x='Gender',data=data)
ax
```

```python
ax = sns.countplot(data = data, x = 'Age Group')
ax
```

```python
ax = sns.countplot(data = data, x = 'Age Group', hue = 'Gender')
ax
```

```python
ax = sns.countplot(data = data, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.barplot(x='Age Group',y='Amount',data=data)
ax
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x='State',y='Amount',data=data)
plt.show()
```

```python
top_state= data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False).head(10)
top_state
plt.figure(figsize=(15,5))
sns.barplot(x='State',y='Amount',data=top_state)
plt.show()
```

```python
ax= sns.countplot(x='Marital_Status',data=data)
ax
```

```python
sales_state= data.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False)
sales_state
ax= sns.barplot(x='Marital_Status', y='Amount',data=sales_state,hue='Gender')
ax
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x='Occupation',y='Amount',data=data)
plt.show()
```

```python
sales_occupation = data.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=sales_occupation, hue='Gender')
```

```python
sales_occupation = data.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=sales_occupation)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(x='Product_Category',data=data)
ax
```

```python
sales_category = data.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_category)
```

```python
sales_ID = data.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID',y='Amount',data=sales_ID)
```
